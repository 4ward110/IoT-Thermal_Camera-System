```python
## Get video name and path
import os
from os import listdir
from os.path import isfile, join

# get folder name
rootdir = '../thermal_video_data'
list_path = []
for file in os.listdir(rootdir):
    d = os.path.join(rootdir, file)
    if os.path.isdir(d):
        print(d)
        list_path.append(str(d))


```

```python
# get files name in specific folder
def get_file_name(folder_path):
    mypath = folder_path
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    print(onlyfiles)
```

```python
# list all files name
for idx, folder_path in enumerate(list_path):
#     get_file_name(folder_path)
    part_of_path = folder_path.split("\\")
    list_path[idx] = '/'.join(part_of_path)
    print(list_path[idx])
```

```python
# get video frames and save as images
import cv2

def get_video_frame(main_folder_path, video_path):
    video_path_split = video_path.split("/")
    main_folder_split = main_folder_path.split("/")
    # create image capture folder path
    parent_dir = "/image_from_video/normal_image"
    if "picam" in video_path_split[-1]:
        image_capture_path = "../image_from_video/normal_image" + "/" + main_folder_split[-1]
        os.mkdir(image_capture_path)
    else:
        image_capture_path = "../image_from_video/thermal_image" + "/" + main_folder_split[-1]
        os.mkdir(image_capture_path)
    
    vidcap = cv2.VideoCapture(video_path)
    success,image = vidcap.read()
    count = 0
    
    while success:
        image_name = "frame"+ str(count) + ".jpg"
        cv2.imwrite(os.path.join(image_capture_path, image_name), image) 
        cv2.waitKey(0)
        success,image = vidcap.read()
        print('Read a new frame: ', success)
        count += 1
```

```python
for path in list_path:
    file_names = os.listdir(path)
    for file in file_names:
        video_path = path + "/" + file
        get_video_frame(path, video_path)
```

```python
# Face detection for thermal image

from PIL import Image
from matplotlib import pyplot as plt
import cv2
import numpy as np
import os
import sys
import time

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# crop face_image
def crop_face(file_name, target_file_name):
    img2 = cv2.imread(file_name)
    face = face_cascade.detectMultiScale(img2,1.1,4)
#     if not face :
#         pass
#     else:
    for (x, y, w, h) in face:
        face_roi = img2[y:y + h, x:x + w]
        cv2.imwrite(target_file_name, face_roi)

    print(" Complete crop: " + file_name + "\n")

# make directory store all processed image
def data_processing(folder_names, path):
    for folder_name in folder_names:
        folder_path = path + folder_name
        file_names = os.listdir(folder_path)
        os.mkdir("../processed_gray_thermal_image/test/" + folder_name)
        for file in file_names:
            image_dir = folder_path + "/" + file
            target_file_name = "../processed_gray_thermal_image/test/" + folder_name + "/" + file
            crop_face(image_dir, target_file_name)
                
```

```python
path = "../train_test_split_data/test/" 
folder_names = os.listdir(path)

# processingz
data_processing(folder_names, path)
```
